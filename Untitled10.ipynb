```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os

# -------------------------
# Chargement de l'image
# -------------------------
def load_image(image_path, channels=3):
    assert os.path.exists(image_path), f"Image not found at {image_path}"
    if channels == 1:
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        assert image is not None, "Failed to load grayscale image"
    elif channels == 3:
        image = cv2.imread(image_path)
        assert image is not None, "Failed to load RGB image"
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    else:
        raise ValueError("Only 1 (grayscale) or 3 (RGB) channels supported")
    return image

# -------------------------
# Application de la convolution
# -------------------------
def apply_convolution(image, kernel):
    assert isinstance(image, np.ndarray), "Image must be a NumPy array"
    assert isinstance(kernel, np.ndarray), "Kernel must be a NumPy array"
    assert len(kernel.shape) == 2 and kernel.shape[0] == kernel.shape[1] and kernel.shape[0] % 2 == 1, \
        "Kernel must be square and have odd size"
    if len(image.shape) == 3:
        assert image.shape[2] == 3, "RGB image must have 3 channels"
        output = np.zeros_like(image, dtype=np.float32)
        for c in range(3):
            output[:, :, c] = convolve_channel(image[:, :, c], kernel)
    else:
        output = convolve_channel(image, kernel)
    return np.clip(output, 0, 255).astype(np.uint8)

def convolve_channel(image, kernel):
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2
    padded = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')
    output = np.zeros_like(image, dtype=np.float32)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            region = padded[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel)
    return output

# -------------------------
# Affichage
# -------------------------
def display_all_filters(original, filtered_images, titles, main_title="Filtered Results"):
    plt.figure(figsize=(15, 8))
    plt.subplot(2, len(filtered_images)+1, 1)
    plt.title("Original")
    plt.imshow(original, cmap='gray' if len(original.shape) == 2 else None)
    plt.axis('off')
    for i, (img, title) in enumerate(zip(filtered_images, titles)):
        plt.subplot(2, len(filtered_images)+1, i+2)
        plt.title(title)
        plt.imshow(img, cmap='gray' if len(img.shape) == 2 else None)
        plt.axis('off')
    plt.suptitle(main_title)
    plt.tight_layout()
    plt.show()

# -------------------------
# Dfinir les filtres de base
# -------------------------
def get_standard_filters():
    blur_kernel = np.ones((3, 3)) / 9
    sobel_horizontal = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_vertical = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    sharpen = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    emboss = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])
    edge_enhance = np.array([[0, 0, 0], [-1, 1, 0], [0, 0, 0]])
    return {
        "Blur": blur_kernel,
        "Sobel Horizontal": sobel_horizontal,
        "Sobel Vertical": sobel_vertical,
        "Sharpen": sharpen,
        "Emboss": emboss,
        "Edge Enhance": edge_enhance
    }

def get_random_kernels(seed=42):
    np.random.seed(seed)
    kernels = {}
    for size in [3, 5, 7]:
        kernel = np.random.randn(size, size)
        kernel /= np.sum(np.abs(kernel)) + 1e-8
        kernels[f"Random {size}x{size}"] = kernel
    return kernels

# -------------------------
# Pipeline principal
# -------------------------
def main():
    image_path = "/content/drive/MyDrive/tp_dl/image.jpg"  #  changer
    gray = load_image(image_path, channels=1)
    rgb = load_image(image_path, channels=3)

    filters = get_standard_filters()
    filters.update(get_random_kernels())

    # Appliquer  limage en niveaux de gris
    gray_results, gray_titles = [], []
    for name, kernel in filters.items():
        filtered = apply_convolution(gray, kernel)
        gray_results.append(filtered)
        gray_titles.append(name)
        cv2.imwrite(f"result_{name.replace(' ', '_').lower()}_gray.jpg", filtered)

    display_all_filters(gray, gray_results, gray_titles, main_title="Gray Image Filters")

    # Appliquer  limage RGB
    rgb_results, rgb_titles = [], []
    for name, kernel in filters.items():
        filtered = apply_convolution(rgb, kernel)
        rgb_results.append(filtered)
        rgb_titles.append(name)
        cv2.imwrite(f"result_{name.replace(' ', '_').lower()}_rgb.jpg", cv2.cvtColor(filtered, cv2.COLOR_RGB2BGR))

    display_all_filters(rgb, rgb_results, rgb_titles, main_title="RGB Image Filters")

if __name__ == "__main__":
    main()
```
